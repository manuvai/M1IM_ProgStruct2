# Principe

Matplotlib est une librairie que utilise la notion de programmation orientée objet (que vous étudierez l'an prochain). Un graphique Matplotlib prend la forme d'un objet nommé ```Figures```, qui va contenir un ou plusieurs ```Axes```, c'est à dire une zone où un ensemble de points peut être spécifié en termes de coordonnées (x,y). 

Matplotlib manipule principalement des données sous la forme de listes. Si l'on travail avec des données csv ou json, il faudra au préalable préparer des fonctions permettant d'extraire les informations avec lesquelles on veut travailler.

Vous pouvez regarder le guide de démarage : [https://matplotlib.org/stable/tutorials/introductory/quick_start.html](https://matplotlib.org/stable/tutorials/introductory/quick_start.html)

Le principe de création d'un graphique repose sur 3 étapes:
1. Création d'une figure à l'aide de la méthode ```fig, ax = plt.subplots()```.
2. Ajout d'un ensemble de données à tracer avec fonction ```ax.plot(x, y)```, où ```x```et ```y``` sont deux listes représentant l'abscisse et l'ordonnée.
3. Affichage du graphique à l'aide de la méthode ```plt.show()```.

Entre l'étape 2 et 3, on va pouvoir préciser des informations concernant l'affichage de notre graphique, comme les labels, ou la couleur des axes. 

Par exemple:

````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()  # Création d'une Figure contenant un seul axe. subplots retourne un couple, une référence vers un objet Figure (fig) et une référence vers un objet axe (ax)
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]);  # On demande à tracer les données passées en paramètre. La première liste correspond aux abscisses et le second aux ordonnées.
plt.show()               # On affiche le graphique 
````

On peut améliorer notre graphique en ajoutant des informations suplémentaires grâce à différentes méthodes:
* ```ax.set_title(str)``` pour ajouter un titre
* ```ax.set_xlabel(str)``` pour ajouter une légende à l'axe des abscisses
* ```ax.set_ylabel(str)``` pour ajouter une légende à l'axe des ordonnées
* ```ax.legend()```pour afficher la légende. Il faut au préalable ajouter un label à notre axe à l'aide du paramètre ```labels``` dans la méthde ```ax.plot```

````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()  # Création d'une Figure contenant un seul axe. subplots retourne un couple, une référence vers un objet Figure (fig) et une référence vers un objet axe (ax)
ax.plot([1, 2, 3, 4], [1, 4, 2, 3], label="Mes données");  # On demande à tracer à l'axe (ax) les données passées en paramètre. La première liste correspond aux abscisses et le second aux ordonnées.
ax.set_title("Mon premier graphique") 
ax.set_xlabel("abscisses")
ax.set_ylabel("ordonnées")
ax.legend()
plt.show()               # On affiche le graphique 
````

Je peux ajouter plusieurs axes dans le même graphique:

````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()  # Création d'une Figure contenant un seul axe. subplots retourne un couple, une référence vers un objet Figure (fig) et une référence vers un objet axe (ax)
ax.plot([1, 2, 3, 4], [1, 4, 2, 3], label="Mes premières données");  # On demande à tracer à l'axe (ax) les données passées en paramètre. La première liste correspond aux abscisses et le second aux ordonnées.
ax.plot([1, 2, 3, 4], [4, 3, 2, 6], label="Mes secondes données");  # On demande à tracer à l'axe (ax) les données passées en paramètre. La première liste correspond aux abscisses et le second aux ordonnées.
ax.set_title("Mon premier graphique") 
ax.set_xlabel("abscisses")
ax.set_ylabel("ordonnées")
ax.legend()
plt.show()               # On affiche le graphique 
````

L'instruction ```fig, ax = plt.subplots()``` permet de créer une figure contenant une seule représentation graphique de données. 

Il est possible d'afficher des graphiques cotes à cotes. L'instruction ```plt.subplots(x,y)``` permet de spécifier les dimensions qu'une grille de taille X x Y. Chaque case de la grille peut alors contenir un graphique.

Par exemple, ```fig, ax = plt.subplots(1, 2)``` créer un figure pour afficher 1 ligne contenant 2 figures. L'accès à une figure se fait grâce à la variable ax qui prend alors la forme d'une liste.

````python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,2)  # Création d'une Figure contenant un seul axe. subplots retourne un couple, une référence vers un objet Figure (fig) et une référence vers un objet axe (ax)

# Première figure
ax[0].plot([1, 2, 3, 4], [1, 4, 2, 3], label="Mes premières données");  # On demande à tracer à l'axe (ax) les données passées en paramètre. La première liste correspond aux abscisses et le second aux ordonnées.
ax[0].set_title("Mon premier graphique") 
ax[0].set_xlabel("abscisses")
ax[0].set_ylabel("ordonnées")
ax[0].legend()

# Deuxième figure : 
ax[1].plot([1, 2, 3, 4], [4, 3, 2, 6], label="Mes secondes données");  # On demande à tracer à l'axe (ax) les données passées en paramètre. La première liste correspond aux abscisses et le second aux ordonnées.
ax[1].set_title("Mon premier graphique") 
ax[1].set_xlabel("abscisses")
ax[1].set_ylabel("ordonnées")
ax[1].legend()

plt.show()               # On affiche le graphique 
````

Il est possible de choisir plusieurs paramètres pour mettre en forme nos graphique. Le site de matplotlib dispose d'une librairie d'exemples: [https://matplotlib.org/stable/gallery/index.html](https://matplotlib.org/stable/gallery/index.html)

# Plot

Ce premier graphique ```plot(x,y)``` permet de tracer des courbes où ```x``` représente les valeurs en abscissees et ```y``` les ordonnées. Matplotlib trace alors une interpolation entre les différents points pour constuire la représentation.

Ce type de graphique est utile lorsque l'on veut observer l'évolution d'une valeure dans le temps. Par exemple, pour tracer l'évolution de la valeure du bitcoin sur les 7 derniers jours:

````python
import matplotlib.pyplot as plt

date = ["19/01/2022","20/01/2022","21/01/2022","22/01/2022","23/01/2022","24/01/2022","25/01/2022"]
valeure= [36800.882813,35957.707031,32135.849609,30877.941406,31986.818359,32369.328125,32693.353516]

fig, ax = plt.subplots()  # Création d'une Figure contenant un seul axe. subplots retourne un couple, une référence vers un objet Figure (fig) et une référence vers un objet axe (ax)
ax.plot(date, valeure, label="BTC-Value");  
ax.set_title("Evolution de la valeure du Bitcoin") 
ax.set_xlabel("Date")
ax.set_ylabel("Valeure en $")
ax.legend()
plt.show()    
````

# Bar

Le diagramme à barres ```bar(x,height)``` permet d'associer à une donnée ```x``` une valeure ```height```. Les données sont alors représentée sous la forme de barres de hauteur ```height```.

C'est un diagramme qui permet de comparer des valeurs. Par exemple, pour comparer le nombre d'étudiants inscrits en lience, master et doctorats en 2020-2021:

````python
import matplotlib.pyplot as plt

categories = ["Licence","Master","Doctorat"]
effectifs = [1014,581,539]

fig, ax = plt.subplots()  # Création d'une Figure contenant un seul axe. subplots retourne un couple, une référence vers un objet Figure (fig) et une référence vers un objet axe (ax)
ax.bar(categories, effectifs);  
ax.set_title("Effectifs LMD en 2021 en milliers d'étudiants") 
plt.show()    
````

# Hist

L'histogramme ```hist(x)``` une représentation graphique permettant de représenter la répartition d'une variable ```x``` avec des colonnes correspondant chacunes à une classe.

Par exemple, il permet de tracer la distribution des notes d'étudiants pour connaitre combiens d'élèves on une note entre 1 et 5, 5 et 10, 10 et 15 et 15 et 20:

````python
import matplotlib.pyplot as plt

notes = [14,12,10,8,3,20,15,15,13,10,8]
classes = [1,5,10,15,20]
fig, ax = plt.subplots()  # Création d'une Figure contenant un seul axe. subplots retourne un couple, une référence vers un objet Figure (fig) et une référence vers un objet axe (ax)

ax.hist(notes,classes);  
ax.set_title("Répartitions des notes") 
plt.show()  
````

# boxplot

La boite à moustaches ```boxplot(x)``` est une représentation graphique permettant de réprésentant sous la forme d'une boîte les informatioins statistiques sur une série de données.

La boite s'étend du premier quartile (Q1) au troisième quartile (Q3) des données, avec une ligne à la médiane. Les moustaches s'étendent à partir de la boîte de 1,5 fois l'écart interquartile (IQR). 

````python
import matplotlib.pyplot as plt

notes = [14,12,10,8,3,20,15,15,13,10,8]
classes = [1,5,10,15,20]
fig, ax = plt.subplots()  # Création d'une Figure contenant un seul axe. subplots retourne un couple, une référence vers un objet Figure (fig) et une référence vers un objet axe (ax)

ax.boxplot(notes);  
ax.set_title("Répartitions des notes") 
plt.show()  
````

# Pie

Le graphique circulaire ou graphique en secteurs (aussi appelé camembert en France) ```pie(x)``` permet de représenter un petit nombre de valeurs (ou de classes) par des angles proportionnels à la fréquence (ou l'effectif) de ces valeurs.

Par exemple, on peut représenter la popularité des langages de programmation:

````python
import matplotlib.pyplot as plt

categories = ["python","java","js","c#","php","c++","R","Objective-c","Swift","Matlab","TypeScript","Ruby"]
popularite = [25.95,21.42,8.26,7.62,7.37,6.31,4.04,3.15,2.56,2.04,1.57,1.53]

fig, ax = plt.subplots()  # Création d'une Figure contenant un seul axe. subplots retourne un couple, une référence vers un objet Figure (fig) et une référence vers un objet axe (ax)
ax.pie(popularite, labels = categories, autopct='%1.1f%%');  
ax.set_title("Poularité des langages informatiques") 
plt.show()  
````
